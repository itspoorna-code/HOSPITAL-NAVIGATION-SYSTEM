```python
import folium
from folium import plugins
import pandas as pd

import ipywidgets
import os
import json

# Initialize the map
map_UMM = folium.Map(location=[12.313092, 76.649574],width="75%", zoom_start=17)

# Add GeoJSON to the map
geojson_path = 'krgeo_resourses/dental.geojson'
folium.GeoJson(geojson_path, name="My Location").add_to(map_UMM)

# Display the map
map_UMM
```

```python
testGeoJson = 'krgeo_resourses/path_from_point1/from_point1_dental.geojson'
```

```python
def switchPosition(coordinate):
  temp = coordinate[0]
  coordinate[0] = coordinate[1]
  coordinate[1] = temp
  return coordinate
```

```python
with open(testGeoJson) as f:
  testWay = json.load(f)
0
for feature in testWay['features']:
    path = feature['geometry']['coordinates']
finalPath = list(map(switchPosition,path))
finalPath
# path = 'krgeo_resourses/path_from_point1/from_point1_dental.geojson'
folium.plugins.AntPath(finalPath).add_to(map_UMM)
map_UMM
     
```
