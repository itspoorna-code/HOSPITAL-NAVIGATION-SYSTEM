```python
import folium
from folium import plugins
import json

# Function to dynamically update paths based on user input
def get_paths(department, point):
    base_path = "krgeo_resourses"
    geojson_path = f"{base_path}/{department}.geojson"  # Update dynamically
    testGeoJson = f"{base_path}/path_from_point{point}/from_point{point}_{department}.geojson"  # Update dynamically
    return geojson_path, testGeoJson

    # Take user input for point
user_input_point = input("Enter the point number (e.g., '1' or '2'): ")

# Take user input for department
user_input_department = input("Enter the department name (e.g., 'xray', 'ent'): ")

# Get updated paths
geojson_path, testGeoJson = get_paths(user_input_department, user_input_point)

# Initialize the map
map_UMM = folium.Map(location=[12.313092, 76.649574], width="75%", zoom_start=17)

# Add GeoJSON polygon to the map
folium.GeoJson(geojson_path, name="My Location").add_to(map_UMM)

# Process the path GeoJSON
def switchPosition(coordinate):
    temp = coordinate[0]
    coordinate[0] = coordinate[1]
    coordinate[1] = temp
    return coordinate

with open(testGeoJson) as f:
    testWay = json.load(f)

finalPath = [
    switchPosition(coord)
    for feature in testWay['features']
    for coord in feature['geometry']['coordinates']
]

# Add the path to the map
folium.plugins.AntPath(finalPath).add_to(map_UMM)

# Display the map (for Jupyter Notebook)
map_UMM
```
